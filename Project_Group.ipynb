```python
data_new.isnull().any()
```

```python
data_new[['sex']] = data_new[['sex']].fillna('ไม่ได้ระบุ')
```

```python
data_new[['Unit']] = data_new[['Unit']].fillna('ปี')
```

```python
data_new[['province_of_onset']] = data_new[['province_of_onset']].fillna('ไม่ได้ระบุ')
```

```python
data_new[['risk']] = data_new[['risk']].fillna('ไม่ได้ระบุ')
```

```python
data_new.isnull().any()
```

```python
data_new.dropna()
```

```python
data_f = data_new.dropna()
```

```python
data_f.isnull().any()
```
