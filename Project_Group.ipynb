#รวมตาราง

```python
data_merged = pd.concat([data_01,data_02,data_03,data_04])
data_merged
```

```python
data_new = data_merged[['No.','announce_date','sex','age','Unit','announce_date','province_of_isolation','risk']]
data_new
```

```python
data_new.isnull().any()
```

```python
data_new[['sex']] = data_new[['sex']].fillna('ไม่ได้ระบุ')
```

```python
age = data_new[['age']].mean() #กำหนดให้ Age เป็นค่าเฉลี่ยของอายุ
data_new[['age']] = data_new[['age']].fillna(age)
```

```python
data_new[['Unit']] = data_new[['Unit']].fillna('ปี')
```

```python
data_new[['province_of_isolation']] = data_new[['province_of_isolation']].fillna('ไม่ได้ระบุ')
```

```python
data_new[['risk']] = data_new[['risk']].fillna('ไม่ได้ระบุ')
```
