```python
data_file_path = os.path.join(path,'pm-20-jan-2021.csv')     # ไฟล์ที่เราต้องการ  
dataHWGP_covid = pd.read_csv(data_file_path)   # .read_csv อ่านไฟล์ที่เป็น csv
```

```python
dataHWGP_covid #แสดงข้อมูลตาราง dataHWGP_covid
```

## ตรวจสอบค่า Missing

```python
dataHWGP_covid.isnull().any() # ตรวจสอบค่า Missing ในตารางข้อมูล dataHWGP_covid 
# ตรงไหนเป็นช่องว่างจะขึ้นว่า True ตรงไหนไม่เป็นช่องว่างจะขึ้นว่า False
```

### แก้ค่า Missing ของคอลัมน์ 'sex'

```python
# ลองดู set ข้อมูลของคอลัมน์ 'sex'
set(dataHWGP_covid['sex']) 
```

```python
# แก้ค่า missing ของคอลัมน์ 'sex' โดยใช้ค่า mode
mode_sex = dataHWGP_covid[['sex']].mode().iloc[0,0]
# เติมค่า mode ลงไปแทนค่า missing
dataHWGP_covid[['sex']] = dataHWGP_covid[['sex']].fillna(mode_sex)
```

```python
dataHWGP_covid.isnull().any() # ตรวจสอบค่า Missing ในตารางข้อมูล dataHWGP_covid
# ดูว่าคอลัมน์ไหนมีค่า missing บ้าง จะเห็นว่าคอลัมน์ 'sex' เป็น 'False' นั่นคือไม่มีค่า Missing แล้ว
```

### แก้ค่า Missing ของคอลัมน์ 'age'

```python
# หา mean ของ 'age' เพื่อแทนค่า mean ของ 'age' ในข้อมูลที่เป็นค่า missing
dataHWGP_covid[['age']].mean()
```

```python
avg_age = dataHWGP_covid[['age']].mean() 
# แทนค่าเฉลี่ยของอายุลงในค่าวาง
dataHWGP_covid[['age']] = dataHWGP_covid[['age']].fillna(avg_age)
```

```python
dataHWGP_covid.isnull().any() # ตรวจสอบค่า Missing ในตารางข้อมูล dataHWGP_covid
# ดูว่าคอลัมน์ไหนมีค่า missing บ้าง จะเห็นว่าคอลัมน์ 'age' เป็น 'False' นั่นคือไม่มีค่า Missing แล้ว
```

### แก้ค่า Missing ของคอลัมน์ 'risk'

```python
set(dataHWGP_covid['risk'])   # ลองดู set ข้อมูลของคอลัมน์ 'risk'
```

```python
dataHWGP_covid[['risk']] = dataHWGP_covid[['risk']].fillna('ไม่มีข้อมูล')  # แทนค่า Missing ของคอลัมน์ 'risk' ด้วย 'ไม่มีข้อมูล'
```

```python
dataHWGP_covid.isnull().any() # ตรวจสอบค่า Missing ในตารางข้อมูล dataHWGP_covid
# ดูว่าคอลัมน์ไหนมีค่า missing บ้าง จะเห็นว่าคอลัมน์ 'risk' เป็น 'False' นั่นคือไม่มีค่า Missing แล้ว
```

## เลือกมาใช้เฉพาะข้อมูลที่เราต้องการ

```python
data_GP1 = dataHWGP_covid[['sex','age','risk']] #เลือกเอามาใช้เฉพาะข้อมูลที่ต้องการ
data_GP1
```

```python
set(data_GP1['risk']) # ลองดู set ข้อมูลของคอลัมน์ 'risk'
```

```python
data_GP1_df1 = data_GP1[data_GP1['risk']=='ไปสถานที่แออัด เช่น งานแฟร์ คอนเสิร์ต'] #เลือกมาเฉพาะ 'risk' ที่เป็น 'ไปสถานที่แออัด เช่น งานแฟร์ คอนเสิร์ต'
data_GP1_df1
```

```python
data_GP1_df2 = data_GP1[data_GP1['risk']=='ศูนย์กักกัน ผู้ต้องกัก'] #เลือกมาเฉพาะ 'risk' ที่เป็น 'ทัณฑสถาน/เรือนจำ'
data_GP1_df2
```

```python
data_GP1_df3 = data_GP1[data_GP1['risk']=='สถานบันเทิง'] #เลือกมาเฉพาะ 'risk' ที่เป็น 'สถานบันเทิง'
data_GP1_df3
```

```python
data = pd.concat([data_GP1_df1,data_GP1_df2,data_GP1_df3]) #รวมตาราง
data
```

```python
# เปลี่ยนคอลัมน์ 'sex' จาก String เป็น ตัวเลข
data['sex'].replace({'ชาย':1,'หญิง':50}, inplace=True)
data
```
